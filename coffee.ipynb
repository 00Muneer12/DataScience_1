```python
plt.figure()
df['transaction_qty'].hist(bins=20)
plt.title("Distribution of Transaction Quantity",c='red')
plt.xlabel("Quantity per Transaction",c='blue')
plt.ylabel("Number of Transactions",c='blue')
plt.show()
```

```python
plt.figure()
df['Total_Bill'].hist(bins=20)
plt.title("Distribution of Total Bill Amount",c='red')
plt.xlabel("Total Bill ($)",c='blue')
plt.ylabel("Number of Transactions",c='blue')
plt.show()
```

```python
hourly_sales = df.groupby('Hour')['Total_Bill'].sum()

plt.figure()
hourly_sales.plot(kind='line')
plt.title("Total Sales by Hour",c='red')
plt.xlabel("Hour of Day",c='blue')
plt.ylabel("Total Sales",c='blue')
plt.show()
```

```python
day_sales = df.groupby('Day Name')['Total_Bill'].sum()

plt.figure()
day_sales.plot(kind='bar')
plt.title("Total Sales by Day of Week",c='red')
plt.xlabel("Day",c='blue')
plt.ylabel("Total Sales",c='blue')
plt.show()
```

```python
category_sales = df.groupby('product_category')['Total_Bill'].sum()

plt.figure()
category_sales.plot(kind='bar')
plt.title("Revenue by Product Category",c='red')
plt.xlabel("Category",c='blue')
plt.ylabel("Total Revenue",c='blue')
plt.show()
```

```python
store_sales = df.groupby('store_location')['Total_Bill'].sum()

plt.figure()
store_sales.plot(kind='bar')
plt.title("Total Revenue by Store Location")
plt.xlabel("Store Location")
plt.ylabel("Total Revenue")
plt.show()
```

```python
store_qty = df.groupby('store_location')['transaction_qty'].sum()

plt.figure()
store_qty.plot(kind='bar')
plt.title("Total Quantity Sold by Store")
plt.xlabel("Store Location")
plt.ylabel("Total Quantity Sold")
plt.show()
```

```python
store_avg_bill = df.groupby('store_location')['Total_Bill'].mean()

plt.figure()
store_avg_bill.plot(kind='bar')
plt.title("Average Bill per Transaction by Store")
plt.xlabel("Store Location")
plt.ylabel("Average Bill")
plt.show()
```

```python
store_transactions = df.groupby('store_location')['transaction_id'].count()

plt.figure()
store_transactions.plot(kind='bar')
plt.title("Number of Transactions by Store")
plt.xlabel("Store Location")
plt.ylabel("Transaction Count")
plt.show()
```

```python
store_performance = df.groupby('store_location').agg(
    Total_Revenue=('Total_Bill', 'sum'),
    Total_Quantity=('transaction_qty', 'sum'),
    Avg_Bill=('Total_Bill', 'mean'),
    Transactions=('transaction_id', 'count')
)

store_performance
```

```python
category_sales = df.groupby('product_category')['Total_Bill'].sum().sort_values(ascending=False)

plt.figure()
category_sales.plot(kind='bar')
plt.title("Total Revenue by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Total Revenue")
plt.show()
```

```python
category_qty = df.groupby('product_category')['transaction_qty'].sum().sort_values(ascending=False)

plt.figure()
category_qty.plot(kind='bar')
plt.title("Total Quantity Sold by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Total Quantity Sold")
plt.show()
```

```python
top_products = (
    df.groupby('product_type')['Total_Bill']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

plt.figure()
top_products.plot(kind='bar')
plt.title("Top 10 Products by Revenue")
plt.xlabel("Product")
plt.ylabel("Total Revenue")
plt.show()
```

```python
top_qty_products = (
    df.groupby('product_type')['transaction_qty']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

plt.figure()
top_qty_products.plot(kind='bar')
plt.title("Top 10 Products by Quantity Sold")
plt.xlabel("Product")
plt.ylabel("Quantity Sold")
plt.show()
```

```python
avg_price = df.groupby('product_type')['unit_price'].mean().sort_values(ascending=False)

plt.figure()
avg_price.plot(kind='bar')
plt.title("Average Unit Price by Product")
plt.xlabel("Product")
plt.ylabel("Average Price")
plt.show()
```

```python
product_performance = df.groupby('product_type').agg(
    Total_Revenue=('Total_Bill', 'sum'),
    Total_Quantity=('transaction_qty', 'sum'),
    Avg_Price=('unit_price', 'mean'),
    Transactions=('transaction_id', 'count')
).sort_values('Total_Revenue', ascending=False)

product_performance.head(10)
```

```python
plt.figure()
plt.scatter(
    product_performance['Total_Quantity'],
    product_performance['Total_Revenue']
)
plt.xlabel("Total Quantity Sold")
plt.ylabel("Total Revenue")
plt.title("High Volume vs High Value Products")
plt.show()
```
